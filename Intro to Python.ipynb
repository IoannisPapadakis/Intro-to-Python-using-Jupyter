### Create a Function

```python
#I'm defining a function that I can recall later
# the blue is the function name followed by arguments (not all have to be required arguments)
def nrm(data, column):
    m = np.mean(column)
    s = np.std(column)
    data['n'] = (column - m)/ s

    '''
    nrm normalizes a column and adds that normalized column to your DataFrame
    
    Parameters
    ----------
    data: dataframe
    column: column in your dataframe
    '''
```

```python
#Check out our function in action (call it anytime in this document; you can also call function through text (like sublime))
nrm(c, c['gini'])
c
# new column is at the end, n
```
